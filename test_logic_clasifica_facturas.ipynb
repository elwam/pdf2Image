```python
# Celda 4: Función para Ejecutar una Prueba Individual contra el Endpoint
def probar_endpoint_clasificacion_factura(nombre_caso: str, texto_documento: str, esperado: int):
    """
    Envía una solicitud al endpoint de clasificación de facturas y verifica el resultado.
    """
    print(f"\n--- Probando Caso API: {nombre_caso} ---")
    # Descomenta la siguiente línea si quieres ver el texto completo de cada caso
    # print(f"Texto de Entrada:\n{texto_documento}\n--------------------")
    print(f"Texto (primeros 80 chars): '{texto_documento[:80].replace('\n', ' ')}...'")
    
    # El payload para el endpoint es un JSON con la clave "texto_documento"
    payload = {"texto_documento": texto_documento}
    
    try:
        response = requests.post(ENDPOINT_URL, json=payload)
        response.raise_for_status()  # Lanza excepción para errores HTTP 4xx/5xx
        
        resultado_api = response.json()
        examenes_facturados_api = resultado_api.get("examenesFacturados")
        decision_source_api = resultado_api.get("decision_source", "N/A") # 'N/A' si la clave no existe
        
        print(f"  Respuesta API (Estado {response.status_code}): {resultado_api}")
        
        if examenes_facturados_api == esperado:
            print(f"  ✅ PRUEBA API PASADA: Esperado '{esperado}', Obtenido '{examenes_facturados_api}' (Fuente: {decision_source_api})")
        else:
            print(f"  ❌ PRUEBA API FALLIDA: Esperado '{esperado}', Obtenido '{examenes_facturados_api}' (Fuente: {decision_source_api})")
            
    except requests.exceptions.ConnectionError:
        print(f"  [ERROR DE CONEXIÓN] No se pudo conectar a la API en {ENDPOINT_URL}.")
        print("  Asegúrate de que el servidor FastAPI (uvicorn) esté ejecutándose.")
    except requests.exceptions.HTTPError as http_err:
        print(f"  [ERROR HTTP: {http_err.response.status_code}] Respuesta del servidor: {http_err.response.text}")
    except json.JSONDecodeError:
        print(f"  [ERROR DE JSON] La respuesta del servidor no es un JSON válido.")
        print(f"  Respuesta recibida: {response.text if response else 'No response object'}")
    except Exception as e:
        print(f"  [ERROR INESPERADO] Ocurrió un error: {e}")
        import traceback
        traceback.print_exc()
```

```python
# Celda 5: Ejecutar todos los Casos de Prueba contra el Endpoint
print("\n=== INICIANDO PRUEBAS DEL ENDPOINT /clasificar-factura-examenes ===")
for caso in casos_de_prueba_facturas_api:
    probar_endpoint_clasificacion_factura(
        nombre_caso=caso["nombre"],
        texto_documento=caso["texto"],
        esperado=caso["esperado_examenesFacturados"]
    )
print("\n=== FIN DE LAS PRUEBAS DEL ENDPOINT DE FACTURAS ===")
```
